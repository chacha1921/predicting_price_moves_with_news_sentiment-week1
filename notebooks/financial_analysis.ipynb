```python
import sys
import os

# Add project root to path
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))

import pandas as pd
from src.finance_utils import (
    load_price_data,
    add_moving_averages,
    add_rsi,
    add_macd,
    plot_price_with_indicators,
    plot_volume,
    plot_correlation_heatmap,
    plot_candlestick
)
```

```python

stocks = {
    "Apple (AAPL)": "../data/yfinance_data/AAPL_historical_data.csv",
    "Amazon (AMZN)": "../data/yfinance_data/AMZN_historical_data.csv",
    "Google (GOOG)": "../data/yfinance_data/GOOG_historical_data.csv",
    "Meta (META)": "../data/yfinance_data/META_historical_data.csv",
    "Microsoft (MSFT)": "../data/yfinance_data/MSFT_historical_data.csv",
    "Nvidia (NVDA)": "../data/yfinance_data/NVDA_historical_data.csv",
    "Tesla (TSLA)": "../data/yfinance_data/TSLA_historical_data.csv"
}
```

```python

# for stock_name, filepath in stocks.items():
#     print(f"Processing {stock_name}...")
#     df = load_price_data(filepath)
#     df = add_moving_averages(df)
#     df = add_rsi(df)
#     df = add_macd(df)
#     plot_price_with_indicators(df, stock_name)
#     output_path = filepath.replace(".csv", "_with_indicators.csv")
#     df.to_csv(output_path, index=False)
#     print(f"Saved processed data to {output_path}\n")


for stock_name, filepath in stocks.items():
    print(f"Processing {stock_name}...")
    df = load_price_data(filepath)
    df = add_moving_averages(df)
    df = add_rsi(df)
    df = add_macd(df)

    plot_price_with_indicators(df, stock_name)
    plot_volume(df, stock_name)
    plot_correlation_heatmap(df, stock_name)
    plot_candlestick(df, stock_name)

    output_path = filepath.replace(".csv", "_with_indicators.csv")
    df.to_csv(output_path, index=False)
    print(f"Saved processed data to {output_path}\n")
```
